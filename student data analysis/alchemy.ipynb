```python
import pandas as pd
from sqlalchemy import create_engine

correct_answer=pd.read_csv(r"C:\Users\shish\OneDrive\Desktop\python class\SQL\project\correct_answer.csv")
question_paper_code=pd.read_csv(r"C:\Users\shish\OneDrive\Desktop\python class\SQL\project\question_paper_code.csv")
student_list=pd.read_csv(r"C:\Users\shish\OneDrive\Desktop\python class\SQL\project\student_list.csv")
student_response=pd.read_csv(r"C:\Users\shish\OneDrive\Desktop\python class\SQL\project\student_response.csv")

DATABASE_URL = "mysql+pymysql://root:shishir123@localhost/student_data"
engine = create_engine(DATABASE_URL)
correct_answer.to_sql(
        name='correct_answer',
        con=engine,
        if_exists='replace',  
        index=False        
    )
question_paper_code.to_sql(
        name='question_paper_code',
        con=engine,
        if_exists='replace',  
        index=False        
    )
student_list.to_sql(
        name='student_list',
        con=engine,
        if_exists='replace',  
        index=False        
    )
student_response.to_sql(
        name='student_response',
        con=engine,
        if_exists='replace',  
        index=False        
    )
```
