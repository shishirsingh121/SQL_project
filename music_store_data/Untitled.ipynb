```python
import pandas as pd
from sqlalchemy import create_engine
import os

# ---------- CONFIGURATION ----------
csv_folder = r"C:\Users\shish\OneDrive\Desktop\python class\SQL\project\music_store_data\music store data"
db_url = "mysql+pymysql://root:shishir123@localhost:3306/music_store"

DATABASE_URL = "mysql+pymysql://root:shishir123@localhost/student_data"
engine = create_engine(db_url)

for file in os.listdir(csv_folder):
    if file.endswith(".csv"):
        file_path = os.path.join(csv_folder, file)
        table_name = os.path.splitext(file)[0]
        print(f"Loading {file} into table '{table_name}'...")
        df = pd.read_csv(file_path)
        df.to_sql(table_name, con=engine, index=False, if_exists="replace")
        
print("######  All CSV files have been imported into the database successfully. ##################")
```
